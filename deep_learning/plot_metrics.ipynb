```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
metrics_path = '/home/spsalmon/towbintools_pipeline/deep_learning/lightning_logs/version_1/metrics.csv'

df = pd.read_csv(metrics_path)

print(df)

print(df['train_loss_epoch'].values.shape)

train_loss = df['train_loss_epoch'].values
# remove nan
train_loss = train_loss[~np.isnan(train_loss)]

val_loss = df['val_loss_epoch'].values
# remove nan
val_loss = val_loss[~np.isnan(val_loss)]

# Plot the loss
plt.plot(train_loss, label='train_loss')
plt.show()
plt.plot(val_loss, label='val_loss')
plt.show()
```
