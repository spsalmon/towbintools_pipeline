# Custom Building Blocks

This block lets you run any alien script as part of the pipeline, and to integrate the results in the filemap. As it's impossible to predict what the output will be, the rerun option for this script is way less smart. If rerun is false, the script won't run if the folder or file it's supposed to output already exists. If rerun is set to true, then everything will be rerun. It is important that the output of your script follows the general scheme of the pipeline. If your script outputs a folder of images, they should contain Time and Point information in their name (specified by **time_regex** and **point_regex** in the config) so that they can be properly linked to the rest of the filemap. If your script outputs a csv, it should contain "Time" and "Point" columns so that the results can be properly joined to the rest of the filemap.

## Options

* **custom_script_path** : The path to your custom script. It can either be a python '.py' or bash '.sh' file.
* **custom_script_name** : The name that will be used for naming the filemap column, the output file / output folder
* **custom_script_return_type** : "subdir" if it outputs a directory of images, "csv" if it outputs a csv (kinda obvious), or null if your script outputs nothing.
* **custom_script_parameters** : A list of parameters that will be added to the command when your script is called.

By default every custom script will be fed the path to the analysis filemap through the --filemap argument, the output path through the --output argument, the block's configuration in pickled form through the --block_config argument, and the global configuration in pickled form through the --config argument, so your custom scripts should have a way to parse them.

### Example

```yaml
custom_script_path: [ "./custom_scripts/pumping_rate.py" ]
custom_script_name: [ "pumping_rate" ]
custom_script_return_type: [ "csv" ]
custom_script_parameters: [ [ "--input_column analysis_movies/raw_movies_str", "--frame_rate 25", "--height 1500", "--std_coeff 1", "--dist 3", "--width 1" ] ]
```

You will find an example of a script measuring pumping rate in high frame rate movies in 'custom_scripts/pumping_rate.py'. You can use this as a template for your own custom scripts.
