# Molt Detection

This block automatically detects the time of molt ends for each individual using extracted time series and a deep learning model. It can be used to align time series later on or rescale them to development progression. This block will make use of any quality control that was run so it's a good idea to run it last. The "legacy" method detects molts by looking for specific patterns in the time series. It is way less accurate than the newer deep learning method but does not need to be trained. The model does not directly detect the hatch point but this block looks at the result of the quality control to do so. 

When molts are detected, we automatically compute every available feature's value at the time of molt end and add them as new columns. Those values are computed by smoothing the time series to remove outliers and then taking the value of the smoothed time series at the time of molt end (for each molt). This simplifies the downstream analysis.

Instructions on how to train the model can be found in the [training section](../training/TrainingMoltDetectionModels.md).

Even though automatic detection is pretty accurate, we recommend to check and correct the results using the GUI. Instructions on how to use the GUI can be found [here](../usage/UsingGUI.md).

## Options

- molt_detection_method : The method used for molt detection **(e.g. [ "deep_learning" ])**. Defaults to **"deep_learning"**.
- molt_detection_model_path : The path to the model used for molt detection **(e.g. [ './models/molt_detection_model.ckpt' ])**. Defaults to **'./models/molt_detection_model.ckpt'**.
- molt_detection_columns : The columns used for molt detection **(e.g. [ [ 'ch2_seg_str_volume' ] ])**.
- molt_detection_batch_size : The batch size used for molt detection **(e.g. [ 1 ])**. Defaults to **1**.

## Example

```yaml
molt_detection_method: [ "deep_learning" ]
molt_detection_model_path: [ './models/molt_detection_model.ckpt' ]
molt_detection_columns: [ [ 'ch2_seg_str_volume' ] ]
molt_detection_batch_size: [ 1 ]
```

This would use the deep learning model located at './models/molt_detection_model.ckpt' to detect molts using the 'ch2_seg_str_volume' column as input, with a batch size of 1. It would add the following columns to the csv file:
* `HatchTime`
* `M1`
* `M2`
* `M3`
* `M4`

For every feature available in the filemap, it would also add a column with the value of that feature at the time of molt end, for example:
* `ch2_seg_str_volume_at_HatchTime`
* `ch2_seg_str_volume_at_M1`
* `ch2_seg_str_volume_at_M2`
* `ch2_seg_str_volume_at_M3`
* `ch2_seg_str_volume_at_M4`
