# Fluorescence Quantification

This block is used for extracting fluorescence statistics within a binary mask. This block returns a structured CSV (or parquet) file containing all the features extracted for each image-mask pair. Different aggregations are possible.
If a background aggregation method is provided, the image will be background substracted before the measurement.

### Available aggregations

- sum
- mean
- median
- max
- min
- standard deviation (std)

### Available background aggregations

- mean
- median
- min

## Options

- fluorescence_quantification_source : The images used for the quantification. The first element of the list is the column of the csv file you want to process, the second element is the channel(s) you want to measure **(e.g. [ 'raw', [ 0 ] ]**)
- fluorescence_quantification_masks : The masks used to determine which pixels to include in the quantification **(e.g. 'analysis/ch1_seg')**
- fluorescence_quantification_aggregations : The aggregation method(s) to use for the measurement(s) **(e.g. [ ['sum'] ])**
- fluorescence_background_aggregation : The statistic used to get the background value **(e.g. [ 'median' ])**. Defaults to **'median'**.

## Example

```yaml
fluorescence_quantification_source: [ [ 'raw', 0 ] ]
fluorescence_quantification_masks: [ 'analysis/ch2_seg' ]
fluorescence_quantification_aggregations: [ ['sum', 'std', 'mean'] ]
fluorescence_background_aggregation: [ 'median' ]
```

This would calculate the sum, standard deviation and mean of the intensities in the first channel of all non-zero pixels of the mask of the second channel, after substracting all the pixels by the median values of all the pixels not within the mask. The new columns would be named:
* `ch1_fluo_sum_on_ch2_seg`
* `ch1_fluo_std_on_ch2_seg`
* `ch1_fluo_mean_on_ch2_seg`
